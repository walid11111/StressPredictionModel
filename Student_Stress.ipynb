# Standarization:
 ### for best model perfomance

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
x=df.drop(['mental_health_history','breathing_problem','blood_pressure','noise_level',
             'living_conditions','safety','teacher_student_relationship',	
              'extracurricular_activities','stress_level'],axis=1)
y=df['stress_level']
X_train, X_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
# Displaying the first few rows before scaling
print("First few rows of X_train before scaling:")
print(X_train.head())
print("First few rows of X_test before scaling:")
print(X_test.head())
```

```python
# scaling the feature
from sklearn.preprocessing import StandardScaler
# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Display the shapes of the scaled data
print(X_train_scaled.shape, X_test_scaled.shape)


# Converting the scaled arrays back to DataFrames for better readability  o)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)  

# Displaying the first few rows after scaling
print("First few rows of X_train after scaling:")
print(X_train_scaled_df.head())
print("First few rows of X_test after scaling:")
print(X_test_scaled_df.head())
```

```python
mean_train = X_train_scaled_df.mean()
std_train = X_train_scaled_df.std()
print("Training Data Mean:\n", mean_train)
print("Training Data Standard Deviation:\n", std_train)

mean_test = X_test_scaled_df.mean()
std_test = X_test_scaled_df.std()
print("Test Data Mean:\n", mean_test)
print("Test Data Standard Deviation:\n", std_test)
```

#  Predictive Modelin:
### I have slect the randomforestclassifier algorithm beacouse our dataset is classification 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Training the Random Forest model for classification
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_scaled_df, y_train)

# Making predictions on the test set
y_pred = model.predict(X_test_scaled_df)

# Evaluating the model using accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

# Displaying the first few predictions
print("First few predictions:")
print(y_pred[:5])
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```
